```python
def load_data(dir):
    img_paths = []

    for root, dirs, files in os.walk(dir):
        for filename in files:
            abs_path = os.path.abspath(os.path.join(root, filename))
            img_paths.append(abs_path)

    return img_paths
```

```python
# Crop images within a directory
img_paths = load_data(dataset_dir)
print('# img paths:', len(img_paths))
print(img_paths)
```

```python
shortened_img_paths = []

for i, img_path in enumerate(img_paths):
    # path_after_dir = r'/'.join(Path(img_path).parts[7:])
    path_after_dir = r'/'.join(Path(img_path).parts[5:])
    shortened_img_paths.append(path_after_dir)
```
